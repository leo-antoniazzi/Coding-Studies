# Git and Github Studies

## Git

- Git is a version control system (VCS) that keeps track of change in files over time.
- It's very important as you can revert changes, go back to previous estates and much more.
- You can colaborate without having to use the same computer.


(Q) Why it is important ?
1. Save your work as snapshots.
2. See what changed and when.
3. Avoid those 'oops' moments where everything breaks.


You can install git using the command:

`brew install git` 

You can then check the version using:

`git --version`


### Basic Git commands

git status
git add my_script.py
git add . 
git commit -m "COMMIT MESSAGE"
git init

How to create a repo

git init
Creates a new .git subdirectory in your current folder ("00 - Coding Studies"), turning it into a Git repository to track changes. This was missing, causing your "not a git repository" error.
​

curl -o .gitignore [URL]
Downloads GitHub's official Python template, creating a .gitignore file. This tells Git to ignore temporary files like .ipynb_checkpoints, pycache, and venv folders common in Jupyter/Python projects.
​

git add .
Stages all files (notebooks, .gitignore) in the current directory for the next commit—prepares them to be saved as a snapshot without actually saving yet.
​

git commit -m "Initial coding studies"
Saves the staged files as a permanent snapshot (version) with a descriptive message. Creates your first commit with full history.
​

git remote add origin git@github.com:leo-antoniazzi/Coding-Studies.git
Links your local repo to the empty private GitHub repo using SSH. "origin" is just a nickname (bookmark) for the full URL—you'll use git push origin instead of typing the whole address every time.

git branch -M main
Renames your default branch from "master" to "main" (GitHub standard). The -M forces it even if a main branch exists locally.
​

git push -u origin main
Uploads your commits to GitHub's "main" branch on the "origin" remote. The -u sets up tracking so future git push/git pull work without specifying origin/main every time. Prompts passphrase once (cached next time).
