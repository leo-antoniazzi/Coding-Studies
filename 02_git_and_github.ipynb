
## ➡️ How to create a repository? 

This is very important as if you don't use Github everyday you may forget...


**Step 1:** 

Every time you want to make it trackable or something you have to initiate the folder as a Git Repository using the command `git init`

- Creates a new .git subdirectory in your current folder ("00 - Coding Studies"), turning it into a Git repository to track changes. This was missing, causing your "not a git repository" error.
​
**Step 2:** 

Stage all files you have in your repository! 

- It prepares the files to the next commit. Not saved just yet! It is important to understand that you can `git add` separate files. But usually you can use `git add .` to save all.

**Step 3:** 

Commit your changes to the repository.
​
- Saves the staged files as a permanent snapshot (version) with a descriptive message using the command `git commit -m "Initial coding studies"`. Creates your first commit with full history.
​
**Step 4:** 

Linking the local repository with the Github repository using SSH.

- `git remote add origin git@github.com:leo-antoniazzi/Coding-Studies.git`
-  The "origin" is just a nickname (bookmark) for the full URL—you'll use git push origin instead of typing the whole address every time.

Note: You can watch this video back to understand the Git workflow better: https://www.youtube.com/watch?v=e9lnsKot_SQ

**Step 5:** 

Renaming to Github Standards

- `git branch -M main`
- Renames your default branch from "master" to "main" (GitHub standard). The -M forces it even if a main branch exists locally.
​
**Step 6:** 

First push and specifying origin/main for next commits

- `git push -u origin main`
- Uploads your commits to GitHub's "main" branch on the "origin" remote. The -u sets up tracking so future git push/git pull work without specifying origin/main every time. Prompts passphrase once (cached next time).


## ➡️ How Git Actually Works?
https://www.youtube.com/watch?v=e9lnsKot_SQ
- Content from ByteByteGo Video on How git Works: Explained in 4 Minutes

The common assumption is that there is only 2 locations where you code is being stored but that's not actually the truth.

There are 4 locations

1. /localWorkingDirectory: Where you change your code. The playground area.
2. Staging Area: Where we store changes before commiting it. 
3. Local Repository: Where we store commited changes locally.
4. Github Remote Repository: Where the code backup is stored for sharing.
