#### 📋 Key Commands

| Command | Purpose | Example |
|---------|---------|---------|
| `pwd` | Show current directory | `pwd` → `/Users/leo/Github/Coding Studies` |
| `ls -la` | List all files (hidden) | Shows `.venv`, `.git` |
| `cd ..` | Go up one directory | `cd ..` → Github folder |
| `mkdir folder` | Create directory | `mkdir notes` |

#### 💡 Key Concepts
- **Absolute paths**: `/Users/leo/folder` (from root)
- **Relative paths**: `./folder` (from current)
- **Hidden files**: Start with `.` (`.venv`, `.gitignore`)

#### 📝 Notes & Gotchas
- Tab completion works everywhere! ⌃
- `cd -` returns to previous directory
- `!!` repeats last command
- `Ctrl+R` searches command history
