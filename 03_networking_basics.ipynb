# 📚 Topic 4: Networking Basics for Software Engineers

#### ➡️ <u>**What happens when you visit a webpage**</u>

1.	DNS Lookup: Your system asks a Domain Name System (DNS) server to translate openai.com into an IP address (e.g., 104.18.10.10).
2.	TCP Connection: Your computer starts a TCP (Transmission Control Protocol) handshake with that IP address.
3.	HTTP Request and Server Response
4. Rendering of page 

#### ➡️ <u>**HTTP v HTTPS**</u>
- HTTP: Unencrypted, data is sent in plain text.
- HTTPS: Secure version using TLS (Transport Layer Security). Encrypts data so it can’t be read or tampered with.

Why it matters:
- HTTPS ensures privacy and data integrity.
- When working with APIs, HTTPS is almost always required.

#### ➡️ <u>**DNS (Domain Name System)**</u>
- Think of DNS like the Internet’s phonebook.
- It translates domain names (e.g. google.com) into IP addresses (e.g. 142.250.64.110).
 
DNS Lookup Flow:
1. Your OS checks local DNS cache.
2. If not found, it queries a DNS server (often provided by your ISP or manually set to Google 8.8.8.8 or Cloudflare 1.1.1.1).

If DNS fails, you’ll see errors like Name or service not known.

#### ➡️ <u>**How VPNs work**</u>
A VPN (Virtual Private Network) routes your traffic through a remote server.

- It encrypts your data and hides your IP address.
- Often used for privacy, remote access, or to bypass location restrictions.

Why this matters:
- VPNs may intercept and modify DNS requests.
- They can also cause cert errors or block services if misconfigured.

#### ➡️ <u>**What is a Firewall**</u>

A Firewall filters incoming and outgoing network traffic.

Types:
- Host-based firewalls (like Windows Defender Firewall)
- Network firewalls (router, corporate level)

What they block:
1. Specific ports (e.g., blocking port 80 or 443)
2. Specific IPs or domains
3. Entire applications (e.g., blocking Python or Jupyter)
