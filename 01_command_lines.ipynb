# Command Line Studies

Putting it simply, the command line is a way to interact with your computer using text-based commands. You can do that using the Terminal.

The computer's files are organized in a tree of folders (directories). The command line gives you tools to explore, navigate and manage these directiories.

* Directory: A folder
* Current Directory: The folder you are currently "in"
* Path: The location of a file or a folder eg. /Users/LeoAntoniazzi/Projects


You have multiple common commands that you may use all the time

* List the Files: `ls`
* Print the Current Directory: `pwd`
* Change Directories: `cd 'folder name'`
    * Change to Home Directory: `cd ~`
    * Change to Parent Directory: `cd ..`
    * Change to Previous Directory: `cd -`
* Show Contents of a file: `cat 'file name'`
* Make a directory: `mkdir new_folder_name`
* Create a file: `touch 'file name'`



It's important to understand that 

Permissions
    r = read
    w = write
    x = execute

Flags
    -a: All = Show all files
    -r: Recursive = Applies the command to a folder and everything in it
    -d: Directory = The directory itself
    -v: Verbose = Makes the terminal talkative showing what is doing
    -h: Human Readable = Changes the PC language to be readable by you
    -f: Force = Make the terminal do something without asking for permission
    -- help

You also have multiple file extensions that you should be aware of:

1. Plain Text: .txt
2. Python Script: .py
3. Markdown: .md
4. Comma-separated Values: .csv
5. JSON Format: .json
6. JSON Line Format: .jsonl


Hidden Files

There are files in your computer that are hidden by default. They are very important and may change things.

They beging with '.' in front of it.

Examples:
1. .gitignore
2. .zshrc

To show all the hidden files in your computer you can just type: 

`ls -a`
