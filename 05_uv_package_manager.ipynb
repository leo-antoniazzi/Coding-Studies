# UV Package Manager - Coding Studies Environment Setup

## What is uv?

**uv** is an extremely fast Python package and project manager written in Rust by Astral. 
It replaces pip + venv + poetry + pip-tools with a single 10-100x faster tool.

**Key Benefits:**
- Lightning-fast package resolution/installation
- Reproducible lockfiles (uv.lock)
- Unified workflow for scripts AND projects
- Drop-in pip/venv compatibility

## Key Commands Used in This Project

| Command | Purpose | Example |
|---------|---------|---------|
| `uv init` | Creates pyproject.toml skeleton | `uv init coding-studies` |
| `uv add pkg` | Add to pyproject.toml + .venv | `uv add langchain openai` |
| `uv sync` | Build .venv from pyproject.toml | `uv sync` |
| `uv pip list` | List installed packages | `uv pip list \| grep ipykernel` |
| `uv run cmd` | Run in project env | `uv run jupyter lab` |

## Complete Environment Setup (Coding Studies)

### 1. Initialize Project
```bash
cd "Coding Studies"
uv init --bare  # Minimal pyproject.toml
```

### 2. Create pyproject.toml
```
[project]
name = "coding-studies"
version = "0.1.0"
description = "AI Engineer Agentic Track - Coding Studies"
dependencies = [
    "langchain>=0.3.0",
    "autogen-agentchat>=0.4.0",
    "openai>=1.0.0",
    "langgraph>=0.2.0",
]

[optional-dependencies]
dev = [
    "jupyter>=1.0.0",
    "ipykernel>=7.0.0",
    "pandas>=2.0.0",
    "matplotlib>=3.8.0",
    "autogen-ext>=0.4.0",
]
```

### 3. Sync the Environment 

uv sync   Creates .venv + uv.lock (exact reproducible versions)
source .venv/bin/activate  (Coding Studies) prompt appears
uv pip list  Verify packages installed

### 3. Sync the Environment 

From activated .venv
python -m ipykernel install --user --name coding-studies-uv --display-name "Coding Studies uv (.venv)"

Verify
>> jupyter kernelspec list  # Should show coding-studies-uv
>> deactivate


# Daily work
source .venv/bin/activate
# Edit notebooks...

# Add new package
uv add pandas  # Updates pyproject.toml + .venv
uv sync        # Lock new versions

# Commit changes
git add pyproject.toml uv.lock *.ipynb
git commit -m "Add pandas analysis"
git push
